```python
import cv2
```

```python
import matplotlib.pyplot as plt
img=cv2.imread(r'F:\Deep Learning Projects\Face detection\istockphoto-1460896518-1024x1024.jpg')
img
```

```python
plt.imshow(img)
```

```python
conv_img=cv2.cvtColor(img,cv2.COLOR_BGR2RGB)
plt.imshow(conv_img)
```

```python
face_cascade=cv2.CascadeClassifier(r'F:\Deep Learning Projects\Face detection\haarcascade_frontalface_default.xml')
face_cascade
```

```python
img_gray=cv2.cvtColor(conv_img,cv2.COLOR_RGB2GRAY)
img_gray
```

```python
face=face_cascade.detectMultiScale(img_gray,minNeighbors=4)
face
```

```python
len(face)
```

```python
for (x,y,w,h) in face:
    cv2.rectangle(conv_img,(x,y),(x+w,y+h),(0,0,255),4)
```

```python
plt.imshow(conv_img)
```

```python
video=cv2.VideoCapture(r'F:\Deep Learning Projects\Face detection\video.mp4')
while True:
    success,frame=video.read()
    gray=cv2.cvtColor(frame,cv2.COLOR_BGR2GRAY)
    faces=face_cascade.detectMultiScale(gray,minNeighbors=12)
    #print(len(faces))
    for (x,y,w,h) in faces:
        cv2.rectangle(frame,(x,y),(x+w,y+h),(0,255,24),3)
    cv2.imshow('VIDEO',frame)
    if cv2.waitKey(1) & 0XFF==ord('q'):
        break
    
video.release()

cv2.destroyAllWindows()    
  
```
