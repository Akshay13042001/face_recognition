```python
import cv2
import matplotlib.pyplot as plt
img=cv2.imread(r'F:\Deep Learning Projects\Face detection\istockphoto-1460896518-1024x1024.jpg')
plt.imshow(img)
```

```python
con_img=cv2.cvtColor(img,cv2.COLOR_BGR2RGB)
plt.imshow(con_img)
```

```python
eye_cascade=cv2.CascadeClassifier(r'F:\Deep Learning Projects\Face detection\haarcascade_eye.xml')
eye_cascade
```

```python
img_gray=cv2.cvtColor(con_img,cv2.COLOR_RGB2GRAY)
img_gray
```

```python
eye=eye_cascade.detectMultiScale(img_gray,minNeighbors=8)
eye
```

```python
len(eye)
```

```python
for (x,y,w,h) in eye:
    cv2.rectangle(con_img,(x,y),(x+w,y+h),(255,0,0),4)
```

```python
#video eye detection
video=cv2.VideoCapture(r'F:\Deep Learning Projects\Face detection\video.mp4')
while True:
    success,frame=video.read()
    gray=cv2.cvtColor(frame,cv2.COLOR_BGR2GRAY)
    faces=eye_cascade.detectMultiScale(gray,minNeighbors=20)
    for (x,y,w,h) in faces:
        cv2.rectangle(frame,(x,y),(x+w,y+h),(0,220,225),3)
    cv2.imshow('EYES',frame)
    if cv2.waitKey(1) & 0XFF==ord('q'):
        break
video.release()
cv2.destroyAllWindows()

```
